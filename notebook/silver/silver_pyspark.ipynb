Now We Will Explore the Data And the Dataframes in the JSON files Pulled from the Upstream Source (S3 bucket)

```python
df.printSchema()
```

```python
df.show(3,truncate=False)
```

```python
df.count()
```

```python
from pyspark.sql.functions import explode

df_exploded = df.select(explode("data").alias("stock"),
                        "date")
```

```python
from pyspark.sql.functions import col
df_select = df_exploded.select( 
    col("stock.ticker").alias("ticker"),
    col("stock.price").alias("price"),
    col("stock.volume").alias("volume"),
    col("stock.last_trade_time").alias("last_trade_time"),
    col("stock.currency").alias("currency"),
    col("stock.day_high").alias("day_high"),
    col("stock.day_low").alias("day_low"),
    col("stock.day_change").alias("day_change"),
    col("date")  # keep ingestion partition
)
```

```python
df_select.describe().show()
```

Adding Data Quality Rules 


```python
df_clean = df_select.filter(
    (col("price").isNotNull()) &
    (col("price") > 0) &
    (col("ticker").isNotNull()) &
    (col("volume").isNotNull()) &
    (col("volume") >= 0)
)
```

Typecasting the Trade Time which is a String to Timestamp

```python
from pyspark.sql.functions import to_timestamp

df_clean = df_clean.withColumn(
    "last_trade_time",
    to_timestamp("last_trade_time")
)
```

Adding the Ingestion Metadata For Traceability

```python
from pyspark.sql.functions import current_timestamp

df_clean = df_clean.withColumn(
    "ingestion_timestamp",
    current_timestamp()
)
```

Adding Dedeuplication Incase of Redundant Ingestion Edge Case

```python
df_clean = df_clean.dropDuplicates([
    "ticker", "last_trade_time"
])
```

Write back to S3 Silver Folder

```python
df_clean.write\
.mode("overwrite")\
.parquet("s3a://wealth-de-poc-astitva/silver/stock_api/")
```

```python
#Transaction Table
df_txn = spark.read.csv("s3a://wealth-de-poc-astitva/bronze/transactions/",header=True,inferSchema=True)
df_txn.show(5)
df_txn.printSchema()
```

Data Quality Rules For this CSV
price > 0
quantity > 0
valid timestamp
valid customer_id (we'll enforce later -> When Joining Transactions with customers, we'll discover the grain and limits of the customer_id)

```python
from pyspark.sql.functions import try_to_timestamp
from pyspark.sql.functions import expr, col

df_txn_clean = df_txn.withColumn(
    "timestamp",
    expr("try_cast(timestamp as timestamp)")
)
df_txn_clean = df_txn_clean.filter(
    (col("price")>0)&
    (col("quantity")>0)&
    col("timestamp").isNotNull() &
    col("txn_type").isin("BUY", "SELL") 
)
df_txn_clean.show(5)
```

```python
df_customer = spark.read.parquet("s3a://wealth-de-poc-astitva/silver/customers/")
```

```python
df_customer = df_customer.withColumnRenamed(
    "load_date", "customer_load_date"
)
```

Referential Intergrity Check between customers and transactions

```python
df_txn_valid = df_txn_clean.join(
    df_customer,
    on="customer_id",
    how="inner"
)
df_txn_valid.show()
```

```python
df_txn_valid.write \
.mode('overwrite')\
.parquet("s3a://wealth-de-poc-astitva/silver/transactions/")
```

```python
df_holdings = spark.read.csv("s3a://wealth-de-poc-astitva/bronze/holdings/",header=True,inferSchema=True)
df_holdings.show(5)
df_holdings.printSchema()
```

```python
from pyspark.sql.functions import col
df_holdings_clean = df_holdings.filter(
(col("quantity") > 0) &
(col("ticker").isNotNull())
)
df_holdings_clean.show(5)
```

```python
df_stock_clean = spark.read.parquet("s3a://wealth-de-poc-astitva/silver/stock_api/")
```

Referential Integrity Check between the join of Holdings and Stocks, Only keeping valid stocks in the holdings

```python
df_holdings_valid = df_holdings_clean.join(
    df_stock_clean.select("ticker").distinct(),
    on="ticker",
    how="inner"
)

df_holdings_valid.show()
```

```python
df_holdings_valid.write\
.mode("overwrite")\
.parquet("s3a://wealth-de-poc-astitva/silver/holdings/")
```
